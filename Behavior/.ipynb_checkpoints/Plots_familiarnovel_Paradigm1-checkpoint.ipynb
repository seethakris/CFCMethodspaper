```python
df1 = pd.read_pickle(os.path.join(FolderName, 'all_combined_Paradigm1B.pkl'))
df1 = df1.iloc[:, 1:]
df2 = pd.read_pickle(os.path.join(FolderName, 'all_combined_Paradigm1A.pkl'))
```

## Freezing by intensity

```python
sub_data = df[df['Paradigm'].str.contains('|'.join(['RecallDay1']))]
columns = [ 'Firstlapfreezing', 'Averagefreezing']
for n, j in enumerate(columns):
    print(j)
    x1 = sub_data[sub_data['VR']=='VR1']
    x2 = sub_data[sub_data['VR']=='VR2']
    for i in sub_data['Shocktype'].unique():
        data1 = x1[x1['Shocktype']==i][j].to_numpy()
        data2 = x2[x2['Shocktype']==i][j].to_numpy()
        print(len(data1))
        print('shock: %s Cohens d: %.3f' % (i, pfs.cohend(data2, data1)))
```

```python
fs, ax = plt.subplots(2, 2, figsize=(5, 5), gridspec_kw={'width_ratios':[2, 1]}, sharey='all')
columns = [ 'Firstlapfreezing', 'Averagefreezing']
hue_order = ['1mA_4', '0.8mA_6', '1mA_6', '1.2mA_6', '0.5mA_12']
sub_data = df[df['Paradigm'].str.contains('|'.join(['RecallDay1']))]
for n, j in enumerate(columns):
    sns.pointplot(data=sub_data, x='VR', hue='Shocktype', y=j, ax=ax[n, 0], palette='Set2', dodge=True, hue_order=hue_order)
    delta_data = sub_data[['Animal', 'Shocktype', 'VR', j]]
    delta_data = delta_data.pivot(columns='VR', index=['Animal', 'Shocktype'], values=j).reset_index()
    delta_data['delta'] = delta_data['VR2'] - delta_data['VR1']
    
#     group_data = sub_data.groupby(by=['Shocktype', 'VR'])[j].mean().reset_index().pivot(columns=['VR'], index='Shocktype').reset_index()
#     group_data['delta'] = group_data[j, 'VR2'] - group_data[j, 'VR1']
#     print(group_data)
    sns.stripplot(data=delta_data, y='delta', hue='Shocktype', ax=ax[n, 1], palette='Set2', marker='s',
                  size=6, jitter=0.4, hue_order=hue_order, alpha=0.8, linewidth=1)

for a in ax[:, 1].flatten():
    #legend
    box = a.get_position()
    a.set_position([box.x0, box.y0, box.width * 0.8, box.height])
    a.legend(loc='center left', bbox_to_anchor=(1, 0.5))

for a in ax[:, 0].flatten():
    a.legend().remove()

#Get no shock baseline
df2 = data[data['Shocktype'].str.contains('no shock')]
sub_data = df2[df2['Paradigm'].str.contains('RecallDay')]
for n, j in enumerate(columns):
    x1 = sub_data[sub_data['VR']=='VR1'][j]
    x2 = sub_data[sub_data['VR']=='VR2'][j]

    ax[n, 0].plot([0, 1], [np.mean(x1), np.mean(x2)], 'k--')
    ax[n, 1].axhline(np.mean(x2)-np.mean(x1), linestyle='--', color='k')
    
sns.despine(fs)
fs.tight_layout()
fs.set_size_inches((4, 4))
fs.savefig(os.path.join(SaveFigureFolder, 'CFCBehavior_byshocktype.pdf'), bbox_inches='tight')
```

## Freezing over days

```python
order = ['RecallDay1', 'RecallDay2', 'RecallDay3', 'RecallDay4']
columns = [ 'Firstlapfreezing', 'Averagefreezing']

sub_data = df[(df['Paradigm'].str.contains('RecallDay'))]
sub_data['RecallDayType'] = sub_data['Paradigm'].str.split('_').str.get(0)

sub_data1 = df[(df['Paradigm'].str.contains('RecallDay')) & (df['VR']=='VR1')]
sub_data2 = df[(df['Paradigm'].str.contains('RecallDay')) & (df['VR']=='VR2')]
sub_data1['RecallDayType'] = sub_data1['Paradigm'].str.split('_').str.get(0)
sub_data2['RecallDayType'] = sub_data2['Paradigm'].str.split('_').str.get(0)

fs, ax = plt.subplots(1, 2, figsize=(5, 3), sharey='all')
for n, c in enumerate(columns):
    print(c)
    sns.pointplot(data=sub_data, x='RecallDayType', y=c, hue='VR', palette='Reds',
              dodge=True, ax=ax[n])
    for n2, ord in enumerate(order):
        data1 = sub_data1[sub_data1['RecallDayType']==ord][c]
        data2 = sub_data2[sub_data2['RecallDayType']==ord][c]

        print('Familiar VR')
        m, ci = pfs.mean_confidence_interval(data1)
        print('%0.2f, %0.2f' %(m, ci))
        print('CFC VR')
        m, ci = pfs.mean_confidence_interval(data2)
        print('%0.2f, %0.2f' %(m, ci))
        t, p = scipy.stats.ttest_rel(data1, data2)
        print('Day: %s, tstst %0.3f, p-value %0.3f' %(ord, t, p))
        ax[n].plot([n2-0.25, n2+0.24], [data1, data2], '.-', color='grey', markersize=4, alpha=0.5) 
sns.despine(fs)
fs.set_size_inches((8, 2))
fs.savefig(os.path.join(SaveFigureFolder, 'CFCBehavior_byrecallday.pdf'), bbox_inches='tight')
```

```python
order = ['Before', 'RecallDay1', 'RecallDay2', 'RecallDay3', 'RecallDay4']
columns = [ 'Firstlapfreezing', 'Averagefreezing']

sub_data1 = df[(df['Paradigm_New'].isin(order)) & (df['VR']=='VR1')]
sub_data2 = df[(df['Paradigm_New'].isin(order)) & (df['VR']=='VR2')]

fs, ax = plt.subplots(1, 2, figsize=(5, 3), sharey='all')
for n, c in enumerate(columns):
    print(c)
    sns.pointplot(data=df[df['Paradigm_New'].isin(order)], x='Paradigm_New', y=c, order=order, hue='VR', palette='Reds',
              dodge=True, ax=ax[n])
    for n2, ord in enumerate(order):
        data1 = sub_data1[sub_data1['Paradigm_New']==ord][c]
        data2 = sub_data2[sub_data2['Paradigm_New']==ord][c]

        print('Familiar VR')
        m, ci = pfs.mean_confidence_interval(data1)
        print('%0.2f, %0.2f' %(m, ci))
        print('CFC VR')
        m, ci = pfs.mean_confidence_interval(data2)
        print('%0.2f, %0.2f' %(m, ci))
        t, p = scipy.stats.ttest_rel(data1, data2)
        print('Day: %s, tstst %0.3f, p-value %0.3f' %(ord, t, p))
        ax[n].plot([n2-0.25, n2+0.24], [data1, data2], '.-', color='grey', markersize=4, alpha=0.5) 
        ax[n].set_ylim((0, 80))
sns.despine(fs)
fs.set_size_inches((10, 2))
fs.savefig(os.path.join(SaveFigureFolder, 'CFCBehavior_byrecallday_FamNov_withbefore.pdf'), bbox_inches='tight')
```

## Epoch

```python
fs, ax = plt.subplots(1, 3, sharex=True, sharey=True)
sub_data = data[(data['Paradigm'].str.contains('RecallDay1')) & (data['VR']=='VR1')]
a = pfs.plot_freezing_bylap(sub_data, ax[0], numlaps=10, legend='Familiar VR', colname='Freezingepoch_bylap')
sub_data = data[(data['Paradigm'].str.contains('RecallDay1')) & (data['VR']=='VR2')]
a = pfs.plot_freezing_bylap(sub_data, ax[0], numlaps=10, legend='CFC VR', colname='Freezingepoch_bylap')

#For Can's use her code
cfc_can = scipy.io.loadmat('/Users/seethakrishnan/Library/CloudStorage/Box-Box/CFCBehavior/Can_laps/by_lap_freeze_count_cfc.mat')
ctrl_can = scipy.io.loadmat('/Users/seethakrishnan/Library/CloudStorage/Box-Box/CFCBehavior/Can_laps/by_lap_freeze_count_ctrl.mat')
numlaps=10
for n, can_data in enumerate([ctrl_can, cfc_can]):
    if n == 0:
        d = can_data['by_lap_freeze_count_task6']
    else:
        d = can_data['by_lap_freeze_count_task7']

    m = []
    ci = []
    for i in range(numlaps):
        this_m, this_ci = pfs.mean_confidence_interval(d[i, :])
        m.append(this_m)
        ci.append(this_ci)
    
    m, ci = np.array(m), np.array(ci)
    ax[1].plot(range(numlaps), m, '.-', markersize=7)
    ax[1].fill_between(range(numlaps), m-ci, m+ci, alpha=0.5)

sub_data = df4[(df4['Paradigm'].str.contains('RecallDay1')) & (df4['VR']=='VR1')]
a = pfs.plot_freezing_bylap(sub_data, ax[2], numlaps=10, legend='Control VR', heather_data=True, colname='Freezingepoch_bylap')
sub_data = df4[(df4['Paradigm'].str.contains('RecallDay1')) & (df4['VR']=='VR2')]
a = pfs.plot_freezing_bylap(sub_data, ax[2], numlaps=10, legend='CFC VR', heather_data=True, colname='Freezingepoch_bylap')

sns.despine(fs)
ax[0].set_title('Fam Nov')
ax[1].set_title('Nov Nov')
ax[2].set_title('Nov Nov Heather')

# for a in ax:
#     a.locator_params(axis='y', nbins=3)
fs.set_size_inches(9, 2)

fs.savefig(os.path.join(SaveFigureFolder, 'Lapwisefreezingepochs.pdf'), bbox_inches='tight')
```
